## Functions

```python
def my_func():
  print("Hello from a function") 
my_func()
```

```python
# Arguments
def my_func(fname):
  print(fname + " CHAN")

my_func("David")
my_func("Peter")
my_func("Liza")
```

```python
def my_func(fname, lname):
  print(fname + " " + lname)

my_func("David", "Chan") 
```

```python
# Keyword arguments
def my_func(student3, student2, student1):
  print("The tallest student is " + student3)

my_func(student1 = "David", student2 = "Peter", student3 = "Liza") 
```

```python
# Default parameter values
def my_func(color = "Red"):
  print("My favorite color is " + color)

my_func("Blue")
my_func("Orange")
my_func()
my_func("Cyan") 
```

```python
# Passing a List argument
def my_func(food):
  for x in food:
    print(x)

fruits = ["apple", "banana", "cherry"]

my_func(fruits)
```

```python
# Return a value
def my_func(x):
  return 5 * x

print(my_func(3))
print(my_func(5))
print(my_func(9)) 
```

```python
# function definitions cannot be empty, but if you for some reason have a function definition with no content, put in the pass statement to avoid getting an error.
def myfunc():
  pass 
```

## Lambda

lambda arguments : expression

```python
x = lambda a : a + 10
print(x(5)) 
```

```python
x = lambda a, b : a * b
print(x(5, 6)) 
```

```python
x = lambda a, b, c : a + b + c
print(x(5, 6, 2)) 
```

## Classes/objects

Python is an object oriented programming language.
Almost everything in Python is an object, with its properties and methods.
A Class is like an object constructor, or a "blueprint" for creating objects.


```python
class MyClass:
  x = 5 
print(MyClass)
print(p1.x)
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("John", 36)

print(p1.name)
print(p1.age) 
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def myfunc(self):
    print("Hello my name is " + self.name)

p1 = Person("John", 36)
p1.myfunc() 
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def myfunc(self):
    print("Hello my name is " + self.name)

p1 = Person("John", 36)
p1.age = 40 

print(p1.age) 
```
