# Tuples

A tuple is a collection which is ordered and unchangeable.

```python
myTuple = ("Arduino", "servo", "sensor")
print(myTuple)
```

```python
myTuple = ("Arduino", "servo", "sensor", “Arduino”)
print(myTuple)
```

```python
# To create a tuple with only one item, a comma must be added after the item, otherwise Python will not recognize it as a tuple.
myTuple1 = ("AI_programming",)
print(type(myTuple1))

#NOT a tuple
myTuple2 = ("Arduino")
print(type(myTuple2))
```

```python
tuple1 = ("car", "ship", "plane")
tuple2 = (10, 20, 30, 40, 50)
tuple3 = (True, True, False)
print(tuple1)
print(tuple2)
print(tuple3)
print(type(tuple1))
print(type(tuple2))
print(type(tuple3))
```

```python
tuple1 = ("female", 1234, True, 4321, "male")
print(tuple1)
```

# While Loops

```python
i = 1
while i < 10:
  print(i)
  i += 1
```

```python
# Break statement
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
# Continue statement
i = 0
while i < 6:
  i += 1 
  if i == 3:
    continue
  print(i)
```

```python
# Else statement
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

# For Loop

```python
sensors = ["temperature sensor", "pressure sensor", "velocity sensor"]
for x in sensors:
  print(x) 
```

```python
for x in "Artificial intelligence":
  print(x) 
```

```python
# Break statement
sensors = ["temperature sensor", "pressure sensor", "velocity sensor"]
for x in sensors:
  print(x) 
  if x == "pressure sensor":
    break
```

```python
# Break
sensors = ["temperature sensor", "pressure sensor", "velocity sensor"]
for x in sensors:
   if x == "pressure sensor":
      break
   print(x)
```

```python
# Continue
sensors = ["temperature sensor", "pressure sensor", "velocity sensor"]
for x in sensors:
  if x == "pressure sensor":
    continue
  print(x) 
```
