## Casting

```python
x = int(1)   # x will be 1
y = int(2.8) # y will be 2
z = int("3") # z will be 3
print(x)
print(y)
print(z)
```

```python
x = float(1)
y = float(2.8)
z = float("3")
w = float("4.2")
print(x)
print(y)
print(z)
print(w)
```

```python
x = str("s1") # x will be 's1'
y = str(2)    # y will be '2'
z = str(3.0)  # z will be '3.0' 
print(x)
print(y)
print(z)
```

Slicing Strings

```python
greeting = "Hello, World!"
print(greeting[:5])
```

```python
greeting = "Hello, World!"
print(greeting[-5:-2])
```

Replace String

```python
greeting = "Hello, World!"
print(greeting.replace("H", "J"))
```

```python
greeting = "Hello, World!"
print(greeting.split(",")) # returns ['Hello', ' World!']
```

String Concatenation

```python
a = "Hello"
b = "World"
c = a + b
print(c)
```

```python
a = "Hello"
b = "World"
c = a + " " + b
print(c)
```

Format - Strings

```python
age = 28
txt = "My name is Peter, and I am {}"
print(txt.format(age))
```

```python
quantity = 3
itemNumber = 567
price = 199.9
myOrder = "I want {} pieces of item {} for {} dollars."
print(myOrder.format(quantity, itemNumber, price))
```

```python
quantity = 3
itemNumber = 567
price = 199.9
myOrder = "I want to pay {2} dollars for {0} pieces of item {1}."
print(myOrder.format(quantity, itemNumber, price))
```
